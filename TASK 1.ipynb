**TASK 1 : Python Basics & Operators**

**1. Python script that accepts two numbers from the user**

```python
x = float(input("Enter the first number: "))
y = float(input("Enter the second number: "))
```

**2. Basic arithmetic operations (add, subtract, multiply, divide, modulus, power)**

```python
# Addition
print(f"Addition: {x} + {y} = {x + y}")

# Subtraction
print(f"Subtraction: {x} - {y} = {x - y}")

# Multiplication
print(f"Multiplication: {x} * {y} = {x * y}")

# Division and Modulus
if y != 0:
    # Division
    print(f"Division: {x} / {y} = {x / y}")

    # Modulus (remainder)
    print(f"Modulus: {x} % {y} = {x % y}")
else:
    print("Division and Modulus: Cannot divide by zero")

# Power operation
print(f"Power: {x} ** {y} = {x ** y}")
```

**3. use of comparison and logical operators**

```python
# Comparison operators
print(f"{x} > {y}  : {x > y}")    # Greater than
print(f"{x} < {y}  : {x < y}")    # Less than
print(f"{x} == {y} : {x == y}")   # Equal to
print(f"{x} != {y} : {x != y}")   # Not equal to


# Logical operators

# AND operator
print(f"({x} > {y}) and ({x} > 0) : {(x > y) and (x > 0)}")

# OR operator
print(f"({x} > {y}) or ({y} > 0)  : {(x > y) or (y > 0)}")

# NOT operator
print(f"not ({x} > {y})          : {not (x > y)}")
```
