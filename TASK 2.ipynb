**TASK 2: Lists and Arrays**

**1. A list of 10 integers using user input**

```python
a = []

print("Enter 10 integers:")

for i in range(10):
    b = int(input(f"Enter number {i + 1}: "))
    a.append(b)

print("\nOriginal List:", a)
```

**2. a) Add an element to the list**

```python
b = int(input("\nEnter a number to add to the list: "))
a.append(b)
print("List after adding element:", a)
```

**2.a) Remove an element from the list**

```python
b = int(input("Enter a number to remove from the list: "))

if b in a:
    a.remove(b)
    print("List after removing element:", a)
else:
    print("Element not found in the list.")
```

**2.b) Find maximum and minimum values**

```python
maximum = max(a)
minimum = min(a)

print(f"\nMaximum value in the list: {maximum}")
print(f"Minimum value in the list: {minimum}")
```

**2.c) Sort the list**

```python
a.sort()
print("\nSorted List:", a)
```

**3. Calculate Mean, Median, and Standard Deviation**

```python
# Mean (average)
mean = sum(a) / len(a)

# Median
n = len(a)
if n % 2 == 0:
    median = (a[n // 2 - 1] + a[n // 2]) / 2
else:
    median = a[n // 2]

# Standard Deviation

# Step 1: Calculate variance
variance = sum((x - mean) ** 2 for x in a) / len(a)

# Step 2: Square root of variance
std_deviation = variance ** 0.5

# Display results
print("\nStatistical Calculations:")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_deviation}")
```

**Lists and Arrays (Using NumPy)**

```python
# Import NumPy library
import numpy as np
```

**1. A list of 10 integers using user input**

```python
a = []

print("Enter 10 integers:")

for i in range(10):
    b = int(input(f"Enter number {i + 1}: "))
    a.append(b)

print("\nOriginal List:", a)
```

**2.a) Add an element to the list**

```python
b = int(input("\nEnter a number to add to the list: "))
a.append(b)
print("List after adding element:", a)
```

```python
b = int(input("Enter a number to remove from the list: "))

if b in a:
    a.remove(b)
    print("List after removing element:", a)
else:
    print("Element not found in the list.")
```

**2.b) Find maximum and minimum values**

**3. Calculate Mean, Median, and Standard Deviation**

```python
# 3.Convert list to NumPy array
array = np.array(a)

# Mean
mean = np.mean(array)

# Median
median = np.median(array)

# Standard Deviation
std_deviation = np.std(array)

# Display results
print("\nStatistical Calculations (Using NumPy):")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_deviation}")
```
