**1.Accepts a user input**

```python
# Import datetime module to get current date and time
from datetime import datetime

# Accept user input (simulating an AI prompt)
prompt = input("Enter your AI prompt: ")

# Get current timestamp
timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
```

**Saved the prompt text in a file named prompt_history.txt with timestamp in append mode**

```python
with open("prompt_history.txt", "a") as f:
    f.write(f"[{timestamp}] {prompt}\n")

# Confirmation message
print("Prompt saved successfully!")
```

```python
with open("prompt_history.txt", "r") as f:
    print("\nPrompt History:\n")

    for line in f:
        print(line.strip())
```
